```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import skew
```

```python
data = pd.read_csv('data.csv')
```

```python
print(data.head())
```

```python
numerical_data = data.select_dtypes(include=['float64', 'int64'])
skewness_values = numerical_data.skew()
print("\nSkewness of Numerical Data:\n", skewness_values)
```

```python
for column in numerical_data.columns:
    plt.figure(figsize=(8, 5))
    sns.histplot(numerical_data[column], kde=True, bins=30)
    plt.title(f'Distribution of {column}')
    plt.xlabel(column)
    plt.ylabel('Frequency')
    plt.show()
```

```python
for column in numerical_data.columns:
    plt.figure(figsize=(8, 5))
    sns.boxplot(x=numerical_data[column])
    plt.title(f'Boxplot of {column}')
    plt.xlabel(column)
    plt.show()
```

```python
print("\nStatistical Method Recommendation:")
for col, skw in skewness_values.items():
    if abs(skw) < 0.5:
        print(f"{col}: Normal distribution - Use Z-Score or other parametric tests.")
    else:
        print(f"{col}: Not normal distribution - Consider non-parametric methods.")
```
